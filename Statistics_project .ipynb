```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# Replace 'diabetes.csv' with your actual file name and path if different
data = pd.read_csv('diabetes.csv')
data = data.drop(columns=['DiabetesPedigreeFunction', 'Pregnancies'])


plt.show()
```

```python

data['BMI'] = data['BMI'].replace(0, data['BMI'].mean())
data['Insulin'] = data['Insulin'].replace(0, data['Insulin'].mean())
data['SkinThickness'] = data['SkinThickness'].replace(0, data['SkinThickness'].mean())
data['BloodPressure'] = data['BloodPressure'].replace(0, data['BloodPressure'].mean())
data['Glucose'] = data['Glucose'].replace(0, data['Glucose'].mean())
```

```python
#wilixikon rank some test
#P-value (1.142e-17): This extremely small p-value (in scientific notation, approximately 0.00000000000000001142) indicates strong evidence against the null hypothesis. It suggests that the observed difference in ranks (or medians) between the two groups is highly unlikely to occur if there were no true difference between the groups' distributions or medians. In most cases, with such a small p-value, the null hypothesis is rejected.

#U Statistic (41950.0): The specific value of the U statistic isn't as straightforward to interpret on its own. It's more useful in comparing to other U statistics or in conjunction with other statistics when assessing effect sizes or differences between groups.

#In summary, the results suggest strong evidence to reject the null hypothesis, indicating that there are significant differences between the distributions of the two groups being compared (possibly their medians, in the context of the Mann-Whitney U test).

from scipy.stats import mannwhitneyu
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'data' is your DataFrame containing the data

# Separate 'Age' data by 'Outcome' groups
age_outcome_0 = data[data['Outcome'] == 0]['Age']
age_outcome_1 = data[data['Outcome'] == 1]['Age']

# Perform Mann-Whitney U test
statistic, p_value = mannwhitneyu(age_outcome_0, age_outcome_1)

print(f"Mann-Whitney U Statistic: {statistic}")
print(f"P-value: {p_value}")

# Boxplot to visualize 'Age' by 'Outcome'
plt.figure(figsize=(8, 6))
sns.boxplot(x='Outcome', y='Age', data=data)
plt.title('Age by Outcome')
plt.xlabel('Outcome')
plt.ylabel('Age')
plt.show()
```

```python
from scipy.stats import ttest_ind

# Perform a t-test for Skin Thickness between individuals with and without diabetes
result_skin_thickness = ttest_ind(data[data['Outcome'] == 1]['SkinThickness'],
                                  data[data['Outcome'] == 0]['SkinThickness'],
                                  equal_var=False)

# Print the test statistic and p-value
print(f"T-test Statistic for Skin Thickness: {result_skin_thickness.statistic}")
print(f"P-value for Skin Thickness: {result_skin_thickness.pvalue}")

# Set the significance level (alpha)
alpha_skin_thickness = 0.05

# Compare p-value to the significance level and provide a conclusion
if result_skin_thickness.pvalue < alpha_skin_thickness:
    print("Reject the null hypothesis. There is evidence of a significant difference in mean Skin Thickness between individuals with diabetes and those without.")
else:
    print("Fail to reject the null hypothesis. There is no significant evidence of a difference in mean Skin Thickness between individuals with diabetes and those without.")
```

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Assuming 'data' is your DataFrame containing the dataset

# Define the dependent variable (Outcome) and independent variables
X = data[[ 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI',  'Age']]
y = data['Outcome']

# Add a constant term to the independent variables matrix (for the intercept)
X = sm.add_constant(X)

# Fit the multiple regression model
model = sm.OLS(y, X).fit()

# Get the model summary
print(model.summary())

# Plotting the predicted vs actual Outcome
predicted = model.predict(X)
plt.scatter(predicted, y)
plt.xlabel('Predicted Outcome')
plt.ylabel('Actual Outcome')
plt.title('Predicted vs Actual Outcome')
plt.show()
```
